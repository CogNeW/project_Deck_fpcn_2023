```python
rest_long_networks = pd.melt(rest , value_vars=['FPCN_BCingulo_Opercular', 'FPCN_BDorsal_Attention','DMN_dorsalFPCN_B', 
                                                'Cingulo_OpercularDorsal_Attention','DMN_dorsalCingulo_Opercular', 'DMN_dorsalDorsal_Attention'])

```

```python
# permutation two-way 1 sample t-test
def t_statistic(x, axis=-1):
    return stats.ttest_1samp(x, popmean=0, axis=axis).statistic
```

```python
sns.set_style('ticks')
sns.despine(trim=True)
```

```python
plt.clf()
plt.figure(figsize=(10, 10))
sns.despine(trim=True)
labels = ['FPCN-B x CON', 'FPCN-B x DAN', 'FPCN-B x dorsal-DMN', 'CON X DAN', 'dorsal-DMN x CON', 'dorsal-DMN x DAN']
ax = sns.barplot(data = rest_long_networks, x = 'variable', y = 'value')
ax.set_ylabel('Pearson R coefficient', fontsize= 25)
ax.set_xlabel('' ,fontsize=25)
ax.set_xticklabels(labels, size = 18, rotation = 20)
ax.tick_params(labelsize=18)
ax.axes.set_title('Between Network Connectivity', fontsize = 35)
plt.tight_layout()
plt.tight_layout()


print(ttest_1samp(rest['FPCN_BCingulo_Opercular'], 0.0))
print(stats.permutation_test((rest['FPCN_BCingulo_Opercular'],), t_statistic, permutation_type='samples'))
print(ttest_1samp(rest['FPCN_BDorsal_Attention'], 0.0))
print(stats.permutation_test((rest['FPCN_BDorsal_Attention'],), t_statistic, permutation_type='samples'))
print(ttest_1samp(rest['DMN_dorsalFPCN_B'], 0.0))
print(stats.permutation_test((rest['DMN_dorsalFPCN_B'],), t_statistic, permutation_type='samples'))
print(ttest_1samp(rest['Cingulo_OpercularDorsal_Attention'], 0.0))
print(stats.permutation_test((rest['Cingulo_OpercularDorsal_Attention'],), t_statistic, permutation_type='samples'))
```
